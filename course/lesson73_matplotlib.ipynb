
# Matplotlib 

Visualisierungen in Python mit Matplotlib:

- Matplotlib ist die Standardbibliothek für Visualisierungen in Python.  
- Man kann mit wenigen Befehlen Diagramme erstellen: Linien, Balken, Scatterplots usw.  
- Häufig wird das Untermodul `pyplot` verwendet (ähnlich wie in MATLAB).


## Grundlagen

````python

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y, label="y = x²")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Grundlegendes Liniendiagramm")
plt.legend()
plt.show()
````

## Liniendiagramm einer Funktion

````python

import numpy as np

x = np.linspace(-5, 5, 100)
y = x**2

plt.plot(x, y, label="y = x²", color="blue")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Quadratfunktion")
plt.legend()
plt.show()
````

## Mehrere Linien im gleichen Diagramm

````python

x = np.linspace(0, 2*np.pi, 200)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin(x)", color="red")
plt.plot(x, y2, label="cos(x)", color="green")
plt.xlabel("x")
plt.ylabel("Wert")
plt.title("Sinus und Cosinus")
plt.legend()
plt.show()
````

## Balkendiagramm

````python

namen = ["Anna", "Ben", "Clara", "David"]
alter = [23, 35, 29, 40]

plt.bar(namen, alter, color="orange")
plt.xlabel("Name")
plt.ylabel("Alter")
plt.title("Alter der Personen")
plt.show()
````

## Histogramm

````python

daten = np.random.randn(1000)  # 1000 normalverteilte Zufallszahlen

plt.hist(daten, bins=30, color="skyblue", edgecolor="black")
plt.title("Histogramm der Daten")
plt.xlabel("Werte")
plt.ylabel("Häufigkeit")
plt.show()
````

## Scatterplot

````python

x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x, y, alpha=0.7, c="red")
plt.title("Scatterplot")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
````

## Subplots

````python

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, axs = plt.subplots(1, 2, figsize=(10, 4))

axs[0].plot(x, y1, "b")
axs[0].set_title("sin(x)")

axs[1].plot(x, y2, "g")
axs[1].set_title("cos(x)")

plt.suptitle("Sinus und Cosinus in Subplots")
plt.show()
````
